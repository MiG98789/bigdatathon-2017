```python
import pandas as pd
import numpy as np
```

```python
with open("kv-data", "r") as fi:
    all_content = fi.read().split("\n")
    
all_content = [[w.replace("\"", "") for w in t.split(",")] for t in all_content]
```

```python
all_content
```

```python
df_dict = []

for content in all_content:
    dic = {}
    for x in content:
        split = x.split(':')
        if (len(split) > 1):
            dic[split[0]] = split[1]
    df_dict.append(dic)    
```

```python
df = pd.DataFrame(df_dict)
```
