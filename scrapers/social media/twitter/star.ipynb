```python
import pandas as pd
import numpy as np
```

```python
with open("tweetssentiment", "r") as fi:
    all_content = fi.read().split("\n")
    
all_content = [[w.replace("\"", "") for w in t.split(",")] for t in all_content]
```

```python
all_content
```

```python
product_id_dict = {}
for content in all_content:
    if len(content) < 2:
        continue

    one = content[0]
    two = content[1]
    
    if one in product_id_dict:
        product_id_dict[one].append({
            "consumer_id": two,
            "sentiment": content[3]
        })
    else:
        product_id_dict[one] = [{
            "consumer_id": two,
            "sentiment": content[3]
        }]
```

```python
df_list = []
for key, value in product_id_dict.items():
    df_dict = {}
    df_dict["product_id"] = key
    for v in value:
        df_dict[v["consumer_id"]] = float(v["sentiment"])
    df_list.append(df_dict)
```

```python
df = pd.DataFrame(df_list)
```

```python
columns = []

for column in df:
    row = df[column]
    score = 0
    for item in row:
        if score > 2:
            columns.append(column)
            break
        if not pd.isnull(item) and not isinstance(item, str):
            if  item > 0:
                score += 1
```

```python
df_a =  df.loc[:, columns]
```

```python
df_a.to_csv('star1.csv', sep=',')
```
